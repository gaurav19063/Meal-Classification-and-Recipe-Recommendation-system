## **Python code to convert scraped JSON dataset to pandas dataframe for efficient processing for classification and recommendation**

```python
import pandas as pd
import json
import re
import numpy as np
```

```python
with open('bbc_food_recipe.json') as json_file: 
    data = json.load(json_file) 
```

```python
print(len(data))
i=0
recipe1=(list(data[1].values())[0])
print(recipe1)
d={}
for i in range(len(data)):
  for x in list(data[i].values())[0]:
    key=list(x.keys())[0]
    if key not in d.keys():
      d[key]=[]
    d[key].append(list(x.values())[0])
df=pd.DataFrame.from_dict(d)
df.head(5)
```

```python
df.to_csv("bbc_food_recipe.csv")
```

```python
df_1 = pd.read_csv("bbc_food_recipe.csv")
df_1.rename(columns={ df_1.columns[0]: "Recipe ID" }, inplace = True)
df_1["Recipe ID"]+=1
ratings=df_1["Rating"].tolist()
norm_ratings_list=[]
num_of_reviews_list=[]
for rating in ratings:
    norm_rating=re.sub("out of ","",rating.strip("[").split(",")[0]).split()[0].strip("'")
    if "None" not in norm_rating:
        norm_rating=round(float(norm_rating)/5,2)
    else:
        norm_rating=None
    norm_ratings_list.append(norm_rating)
    num_of_reviews=int(re.sub("'","",re.sub("reviews","",rating.strip("[").split(",")[1]).strip("]").strip()))
    num_of_reviews_list.append(num_of_reviews)
max_reviews=max(num_of_reviews_list)
num_of_reviews_list=[num_of_reviews/max_reviews for num_of_reviews in num_of_reviews_list]
rating_score_list=[]
for i in range(len(norm_ratings_list)):
    if str(norm_ratings_list[i])!="None":
        rating_score_list.append(round(0.8*norm_ratings_list[i]+0.2*num_of_reviews_list[i],2))
    else:
        rating_score_list.append(None)
df_1=df_1.fillna("0")
df_1["Rating Score"]=rating_score_list
df_1.head()
```

```python
df_2 = pd.read_csv("spark_people_food_recipe.csv")
df_2.rename(columns={ df_2.columns[0]: "Recipe ID" }, inplace = True)
df_2["Recipe ID"]=df_2["Recipe ID"]+len(df_1)+1
df_2.loc[df_2['Meal'] == "breakfast", 'Meal'] = "Breakfast"
df_2.loc[df_2['Meal'] == "lunch", 'Meal'] = "Lunch"
df_2.loc[df_2['Meal'] == "dinner", 'Meal'] = "Dinner"
ratings=df_2["Rating"].tolist()
norm_ratings_list=[]
num_of_reviews_list=[]
for rating in ratings:
    norm_rating=re.sub("out of ","",rating.strip("[").split(",")[0]).split()[0].strip("'")
    if "None" not in norm_rating:
        norm_rating=round(float(norm_rating)/5,2)
    else:
        norm_rating=None
    norm_ratings_list.append(norm_rating)
    norm_num_of_reviews=re.sub("'","",re.sub("reviews","",rating.strip("[").split(",")[1]).strip("]").strip())
    if "None" not in norm_num_of_reviews:
        num_of_reviews=int(norm_num_of_reviews)
    else:
        num_of_reviews=None
    num_of_reviews_list.append(num_of_reviews)

max_reviews=np.nanmax(np.array(num_of_reviews_list, dtype=np.float64))
num_of_reviews_list=[num_of_reviews/max_reviews if num_of_reviews is not None else None for num_of_reviews in num_of_reviews_list]

rating_score_list=[]
for i in range(len(norm_ratings_list)):
    if str(norm_ratings_list[i])!="None":
        rating_score_list.append(round(0.8*norm_ratings_list[i]+0.2*num_of_reviews_list[i],2))
    else:
        rating_score_list.append(None)

max_rating=np.nanmax(np.array(rating_score_list, dtype=np.float64))
rating_score_list=[round(rating_score/max_rating,2) if rating_score is not None else None for rating_score in rating_score_list]

df_2["Cholesterol"]=round(df_2[df_2.Cholesterol.str.match(".*mg")==True]["Cholesterol"].str.strip('mg').str.strip().str.replace(',', '').astype(float)/1000,3)
df_2["Cholesterol"]=df_2["Cholesterol"].astype(str)+" g"


df_2=df_2.fillna("0")
df_2["Rating Score"]=rating_score_list
df_2.head()
```

```python
df=df_1.append(df_2, ignore_index=True)
df.rename(columns={ df.columns[5]: "Preperation Time" }, inplace = True)
df[['Preperation Time', 'Cooking Time']] = df[['Preperation Time', 'Cooking Time']].apply(pd.to_numeric)

df["Fat"]=df.Fat.str.strip('g').str.strip()
df["Cholesterol"]=df.Cholesterol.str.strip('g').str.strip()
df["Carbs"]=df.Carbs.str.strip('g').str.strip()
df["Fiber"]=df.Fiber.str.strip('g').str.strip()
df["Protein"]=df.Protein.str.strip('g').str.strip()

recipes=[]
generic_ingredients_list=[]
with open('train.json','r') as f: 
 recipes = json.load(f)
for recipe in recipes: 
   temp_basket=[item.lower() for item in recipe['ingredients']] 
   generic_ingredients_list+=temp_basket 
generic_ingredients_list=list(set(generic_ingredients_list))

ingredient_lists=list(df["Ingredients"])
lookup_list=[]
for i,ingredient_list in enumerate(ingredient_lists):
    temp_list=[]
    ingredients=re.sub("'","",ingredient_list.strip('[').strip(']')).split(",")
    for ingredient in ingredients:
        ingredient_search_list=ingredient.split()
        ingredient_search_list=[ing.lower() for ing in ingredient_search_list]
        for generic_ingredient in generic_ingredients_list:
            if generic_ingredient in ingredient_search_list:
                temp_list.append(generic_ingredient)
                break
    temp_list=list(set(temp_list))
    lookup_list.append(temp_list)

df["Lookup Ingredients"]=lookup_list

df.to_csv("recipes.csv",index=False)
```
